```python
import pandas as pd
```

# CRISPRi

```python
tabs = pd.read_excel('../data/aah7111-tables9.xls', sheet_name=None)
```

```python
interactions = set()
for k in tabs.keys():
    lncRNA_name = '_'.join(k.split('_')[1:])
    target_names = tabs[k]['gname']
    for t in target_names:
        interactions.add((lncRNA_name, 'affects', t))
        interactions.add((t, 'is_affected_by', lncRNA_name))
```

```python
tabs['u87_AC016831.7']
```

```python
external_val_df = pd.DataFrame(list(interactions), columns=['head', 'relation', 'tail'])
```

```python
external_val_df.to_csv('../data/external_val_crispri.tsv', sep='\t', index=False)
```

# ChIA-PET

```python
chia_pet_df = pd.read_excel('../data/12859_2019_2900_MOESM2_ESM.xlsx', sheet_name='supplamentary_table')
```

```python
cluster_strings = []
for i in range(0, 12):
    cluster_name = 'Cluster' + str(i)
    cluster_strings += chia_pet_df[cluster_name].dropna().tolist()
```

```python
interactions = set()
for c_string in cluster_strings:
    # Remove whitespace and split the string by semicolon
    c_string = c_string.replace(' ', '')
    c_string = c_string.replace('/', ',')
    split_data = c_string.split(';')

    # Extract gene and lncRNA sections
    genes_section = split_data[0]
    lncrnas_section = split_data[1]

    # Extract the gene names
    genes_list = genes_section.split(':')[1].split(',')

    # Extract the lncRNA names
    lncrnas_list = lncrnas_section.split(':')[1].split(',')

    # Add the interactions to the set
    for g in genes_list:
        for l in lncrnas_list:
            if not l == 'none' and not g == 'none':
                interactions.add((g, 'is_affected_by', l))
                interactions.add((l, 'affects', g))

```

```python
interactions
```

```python
external_val_df = pd.DataFrame(list(interactions), columns=['head', 'relation', 'tail'])
```

```python
external_val_df.to_csv('../data/external_val_chia-pet.tsv', sep='\t', index=False)
```
