This notebook contains code for splitting the tuples in `data/dataset.tsv` (generated by `download.ipynb`) into train, val and, test splits. These are stored to `data/train.tsv`, `data/val.tsv`, and `data/test.tsv` respectively.

```python
from dataset import load_split
import pandas as pd
import torch
from torch.utils.data import random_split
```

```python
SPLITS = {"train": .8, "val": .1, "test": .1}
```

load data

```python
dataset = load_split("full").to_dict("records") # extract list[dict]
```

create and store splits

```python
g = torch.Generator().manual_seed(42) # reproducible splits
splits = {ty: pd.DataFrame(list(split)) for ty, split in zip(
  SPLITS.keys(), random_split(dataset, SPLITS.values(), generator=g)
)}
```

```python
for ty, split in splits.items():
  split.to_csv(f"data/{ty}.tsv", sep="\t", index=False)
```
